# Sea Level Projection - Datasheet Analysis

Charting sea level projections from 2005 to 2150

```python
import csv
from slp import Projection

projections = []

file = csv.DictReader(open('slp_data.csv'))
for row in file:
    p = Projection(
        row['PSMSL Site'],
        row['PSMSL ID'],
        row['NOAA ID'],
        row['NOAA Name'],
        row['RSL GridNum'],
        row['Lat'],
        row['Long'],
        row['RSL2005 (cm)'],
        row['RSL2020 (cm)'],
        row['RSL2030 (cm)'],
        row['RSL2040 (cm)'],
        row['RSL2050 (cm)'],
        row['RSL2060 (cm)'],
        row['RSL2070 (cm)'],
        row['RSL2080 (cm)'],
        row['RSL2090 (cm)'],
        row['RSL2100 (cm)'],
        row['RSL2110 (cm)'],
        row['RSL2120 (cm)'],
        row['RSL2130 (cm)'],
        row['RSL2140 (cm)'],
        row['RSL2150 (cm)']
    )
    projections.append(p)

print( projections[0] )
```

```python
import matplotlib.pyplot as plt
import numpy as np
import uuid

plt.style.use('_mpl-gallery')

def draw_graph(projection):
       # make data:
       x = 0.5 + np.arange(6)
       y = [ 
              projection.rsl2005, 
              projection.rsl2020, 
              projection.rsl2030, 
              projection.rsl2040,
              projection.rsl2050,
              projection.rsl2060
       ]

       # build the plot
       fig, ax = plt.subplots()

       ax.bar(x, y, width=1, edgecolor="white", linewidth=0.5)

       ax.set(xlim=(0, 15), xticks=np.arange(1, 15),
              ylim=(0, 30), yticks=np.arange(1, 30))

       plt.xlabel('year')
       plt.ylabel('projected (cm)')
       id = str(uuid.uuid4())
       plt.savefig(f'{projection.noaa_name}_{id}.jpg', bbox_inches='tight', dpi=200)
       print(f'{projection.noaa_name}_{id}.jpg ...saved')

       # plt.show()

for p in projections:
       print(f'projection: {p.lat, p.long}')
       draw_graph(p)
```
